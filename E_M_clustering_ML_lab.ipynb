## ii. E-M clustering

**Expectation-Maximization** - Similar to K-means but instead of assigning examples to clusters to maximize the differences in means for continuous variables, the EM clustering algorithm computes **probabilities** of cluster memberships based on one or more probability distributions. The goal of the clustering algorithm then is to maximize the overall probability or likelihood of the data, given the (final) clusters.

```python
from sklearn import datasets
```

```python
# try clustering the iris dataset
iris = datasets.load_iris()
```

```python
# select only petal length and petal width features
X = iris.data[:,2:]
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3, random_state=0)
```

```python
the_labels = gmm.fit_predict(X)
```

```python
print("The means of each component: ",gmm.means_)
print("Has the model converged: ",gmm.converged_)
print("The weights of each component: ", gmm.weights_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,10))
# the em clustering does not compute centroids
plt.scatter(X[:, 0], X[:, 1], c = the_labels, s=100)
plt.title("The original scatter plot")
plt.xlabel("Petal Length")
plt.ylabel("Petal width")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X[:,0], X[:,1], c = iris.target, s=100)
plt.title("The original scatter plot")
plt.xlabel("Petal Length")
plt.ylabel("Petal width")
cb = plt.colorbar()
import numpy as np
loc = np.arange(0,3,8/float(8))
cb.set_ticks(loc)
cb.set_ticklabels(["Iris1","Iris2","Iris3"])
plt.show()
```
